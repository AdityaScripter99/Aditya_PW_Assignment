```python
# Q15. Decorators in Python:-

#A function that takes another function and extends its behavior without explicitly modifying it.

#Example:
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")


say_hello()    
```

```python
# Q19. Iterators, Iterable, Generators in Python:-

#Iterator: Object that allows you to traverse through all the elements of a collection.
#Iterable: Object that returns an iterator.
#Generator: A special iterator that yields items instead of returning them.

#Example:

#Iterators 

# Create an iterator from a list (iterable)
my_iterator = iter(my_list)

# Access elements one by one using next()
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2

#Iterable

# Example of an iterable (a list)
my_list = [1, 2, 3, 4]

# You can loop through the iterable
for item in my_list:
    print(item)


#Generators

# Example of a generator function
def my_generator():
    yield 1
    yield 2
    yield 3

# Using the generator
gen = my_generator()

for value in gen:
    print(value)
```

```python
# Q20. Difference between xrange and range in Python:-

#range:Generates a list containing all the numbers in the specified range.
#This can consume a lot of memory if the range is large.

#xrange: Returns an iterator that generates the numbers in the range on demand,
#one at a time, without storing the entire list in memory. This is more memory-efficient for large ranges.

#Example:

# Using range in Python 3
numbers = range(1, 10)
print(numbers)  # Output: range(1, 10)

# Convert to a list if you need a list
numbers_list = list(numbers)
print(numbers_list)  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]

# You can loop through it directly
for num in numbers:
    print(num)
```

```python
# Q21. Pillars Of Oops :- 

# The four main pillars of Object-Oriented Programming (OOP) are:-

# 1.Encapsulation:- Encapsulation is the concept of bundling data (attributes) and methods (functions)
# that operate on the data into a single unit, or class. It restricts direct access to some of the 
# object's components, which is a way of preventing accidental interference and misuse of the data.

class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute
        self.__age = age    # Private attribute

    def get_info(self):
        return f"Name: {self.__name}, Age: {self.__age}"

# Creating an object
person = Person("Alice", 30)
print(person.get_info())  # Accessing data via a method


# 2.Inheritance:-Inheritance allows a class (child class) to inherit attributes and methods from another class (parent class). 
# This promotes code reuse and establishes a relationship between classes.

class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an object of the derived class
dog = Dog()
dog.sound()  # Inherited method
dog.bark()   # Child class method


# 3. Polymorphism:-Polymorphism allows objects of different classes to be treated as objects of a common superclass.
# It also allows for the same operation to behave differently on different classes.

class Animal:
    def make_sound(self):
        print("Animal sound")

class Dog(Animal):
    def make_sound(self):
        print("Bark")

class Cat(Animal):
    def make_sound(self):
        print("Meow")

# Polymorphism in action
def animal_sound(animal):
    animal.make_sound()

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: Bark
animal_sound(cat)  # Output: Meow


# 4. Abstraction:-Abstraction is the concept of hiding the complex implementation details of an object and exposing only the essential features.
#It helps in reducing complexity and allows the programmer to focus on interactions at a higher level.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

# Creating objects
dog = Dog()
cat = Cat()

dog.sound()  # Output: Bark
cat.sound()  # Output: Meow


# Summary:-

# Encapsulation: Protects data and restricts access to it.
# Inheritance: Enables code reuse by allowing one class to inherit from another.
# Polymorphism: Allows methods to do different things based on the object it is acting upon.
# Abstraction: Hides complex details and shows only the essentials.

```

```python
# Q23. Inheritance in Python:-

#Single Inheritance: Inheriting from one class.
#Multiple Inheritance: Inheriting from multiple classes.
#Multilevel Inheritance: Inheriting from a derived class.
#Hierarchical Inheritance: Multiple classes inherit from a single class.
#Hybrid Inheritance: A combination of different types of inheritance.

#Example:

#Single Inheritance 

# Base class (Parent)
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Derived class (Child)
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Create an object of the derived class
dog = Dog()
dog.sound()  # Inherited method
dog.bark()   # Child class method


#Multiple Inheritance 

# Base class 1
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Base class 2
class Pet:
    def be_friendly(self):
        print("Pet is friendly")

# Derived class
class Dog(Animal, Pet):
    def bark(self):
        print("Dog barks")

# Create an object of the derived class
dog = Dog()
dog.sound()       # Inherited from Animal
dog.be_friendly() # Inherited from Pet
dog.bark()        # Child class method


#Multilevel Inheritance 

# Base class
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Intermediate class
class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

# Derived class
class Dog(Mammal):
    def bark(self):
        print("Dog barks")

# Create an object of the derived class
dog = Dog()
dog.sound()  # Inherited from Animal
dog.walk()   # Inherited from Mammal
dog.bark()   # Child class method


#Hierarchical Inheritance

# Base class
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Derived class 1
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Derived class 2
class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Create objects of the derived classes
dog = Dog()
cat = Cat()

dog.sound()  # Inherited from Animal
dog.bark()   # Dog's method

cat.sound()  # Inherited from Animal
cat.meow()   # Cat's method


#Hybrid Inheritance

# Base class
class Animal:
    def sound(self):
        print("Animal makes a sound")

# Intermediate class 1
class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

# Intermediate class 2
class Pet:
    def be_friendly(self):
        print("Pet is friendly")

# Derived class
class Dog(Mammal, Pet):
    def bark(self):
        print("Dog barks")

# Create an object of the derived class
dog = Dog()
dog.sound()       # Inherited from Animal
dog.walk()        # Inherited from Mammal
dog.be_friendly() # Inherited from Pet
dog.bark()        # Dog's method


# Summary:-

# Single Inheritance: One parent, one child.
# Multiple Inheritance: Multiple parents, one child.
# Multilevel Inheritance: Inheritance chain, parent → child → grandchild.
# Hierarchical Inheritance: One parent, multiple children.
# Hybrid Inheritance: Combination of two or more types of inheritance.

      
```
