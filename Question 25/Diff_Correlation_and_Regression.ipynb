## Discuss the 4 differences between correlation and regression.

```python
# Answer:

# 1. Purpose:

# Correlation: Measures the strength and direction of a linear relationship between two variables.
# Regression: Describes how one variable can predict another and quantifies the relationship between the dependent and independent variables.

# 2. Symmetry:

# Correlation: Is symmetric, meaning the correlation between X and Y is the same as Y and X.
# Regression: Is not symmetric, meaning the regression of Y on X is different from the regression of X on Y.

# 3. Dependence:

# Correlation: Does not imply causality; it only measures the degree of association between variables.
# Regression: Implies causality, where one variable (independent) is used to predict or explain another variable (dependent).

# 4. Equation:

# Correlation: Uses the Pearson correlation coefficient formula which has no concept of dependent or independent variables.
# Regression: Results in an equation of the form  ð‘Œ=ð‘Ž+ð‘ð‘‹ , where b is the slope (regression coefficient) and a is the intercept.
```

```python
import numpy as np
from scipy.stats import pearsonr, linregress
import matplotlib.pyplot as plt

def calculate_correlation(x, y):
    # Calculate Pearson correlation coefficient
    correlation_coefficient, _ = pearsonr(x, y)
    return correlation_coefficient

def calculate_regression(x, y):
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    return slope, intercept, r_value

# Example usage
x = np.array([10, 20, 30, 40, 50])
y = np.array([12, 24, 33, 45, 53])

# Calculate correlation
correlation = calculate_correlation(x, y)
print(f"Pearson Correlation Coefficient: {correlation}")

# Calculate regression
slope, intercept, r_value = calculate_regression(x, y)
print(f"Linear Regression Equation: y = {slope:.2f}x + {intercept:.2f}")
print(f"R-squared: {r_value**2}")

# Plot the data and regression line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Correlation and Regression')
plt.legend()
plt.show()
```
