### What do you mean by skewness? Explain its types. Use a graph to show.

```python
# Answer:

#Skewness: Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. 
#It can tell you about the direction and relative magnitude of a distribution's deviation from the normal distribution.

# Types of Skewness:

# Positive Skewness (Right Skew): The tail on the right side of the distribution is longer or fatter. Most of the data points lie to the left of the mean.
# Negative Skewness (Left Skew): The tail on the left side is longer or fatter. Most of the data points lie to the right of the mean.
# Zero Skewness (Symmetric Distribution): When a distribution is perfectly symmetrical, it has zero skewness.

# Graphical Representation:

# Positive Skew: The peak is on the left, tail extends to the right.
# Negative Skew: The peak is on the right, tail extends to the left.
# Zero Skew: A bell-shaped curve.
```

```python
from scipy.stats import skew

def calculate_skewness(data):
    return skew(data)

# Example usage
data = [10, 15, 14, 14, 18, 16, 15, 16, 14]
skewness = calculate_skewness(data)
print(f"Skewness: {skewness}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data for different types of skewness
data_symmetrical = np.random.normal(0, 1, 1000)  # Symmetrical (normal) distribution
data_positive_skew = np.random.exponential(1, 1000)  # Right-skewed
data_negative_skew = -np.random.exponential(1, 1000)  # Left-skewed

# Plotting the distributions
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(data_symmetrical, kde=True, ax=axs[0], color='blue')
axs[0].set_title('Symmetrical Distribution')

sns.histplot(data_positive_skew, kde=True, ax=axs[1], color='green')
axs[1].set_title('Right-Skewed Distribution')

sns.histplot(data_negative_skew, kde=True, ax=axs[2], color='red')
axs[2].set_title('Left-Skewed Distribution')

plt.show()
```
